```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

# Increased training data
celsius = np.array([
    -100, -80, -60, -40, -20, -10, 0, 5, 8, 10,
    15, 20, 22, 25, 30, 38, 40, 50, 60, 80, 100
], dtype=float)

fahrenheit = np.array([
    -148, -112, -76, -40, -4, 14, 32, 41, 46, 50,
    59, 68, 72, 77, 86, 100, 104, 122, 140, 176, 212
], dtype=float)

# Model
model = keras.Sequential([
    keras.Input(shape=(1,)),
    keras.layers.Dense(units=1)
])

# Compile
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.1),
    loss='mean_squared_error'
)

# Increased epochs
model.fit(celsius, fahrenheit, epochs=3000, verbose=0)

# Test prediction
test_value = np.array([[100.0]])
prediction = model.predict(test_value)

print(f"{test_value}°C = {prediction[0][0]:.2f}°F")
```

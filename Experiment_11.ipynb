```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

texts = [
    "Natural Language Processing is very interesting and useful.",
    "NLTK is a powerful library for text processing.",
    "SpaCy is fast and efficient for NLP tasks.",
    "Machine learning models require clean and processed data.",
    "Tokenization, stemming, and lemmatization are important steps in NLP."
]
# Download required resources
nltk.download('punkt')

ps = PorterStemmer()

for sentence in texts:
    tokens = word_tokenize(sentence)
    stemmed = [ps.stem(word) for word in tokens]
    
    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Stemmed:", stemmed)
    print()
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download resources
nltk.download('stopwords')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

for sentence in texts:
    tokens = word_tokenize(sentence)
    
    filtered = [word for word in tokens if word.lower() not in stop_words]
    lemmatized = [lemmatizer.lemmatize(word) for word in filtered]
    
    print("Original:", sentence)
    print("After Stopword Removal:", filtered)
    print("Lemmatized:", lemmatized)
    print()
```

```python
import spacy

# Load model
nlp = spacy.load("en_core_web_sm")

for sentence in texts:
    doc = nlp(sentence)
    
    processed = [
        token.lemma_ for token in doc 
        if not token.is_stop and not token.is_punct
    ]
    
    print("Original:", sentence)
    print("Processed (SpaCy):", processed)
    print()
```
